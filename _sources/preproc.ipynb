Tujuan tahap ini adalah menghapus outlier agar model belajar dari pola data yang lebih konsisten dan akurat.


## **Menangani Outlier**

Outlier perlu diperhatikan saat preprocessing karena bisa mengganggu analisis. Nilai ekstrem dapat menggeser statistik seperti rata-rata dan standar deviasi, serta membuat model salah mengenali pola. Dalam deteksi anomali, outlier biasanya ditandai dengan Anomaly = 1.
Cara umum menangani outlier antara lain:
- Menghapus — cocok jika jumlahnya kecil dan tidak mewakili populasi.
- Transformasi — seperti log atau z-score, jika outlier tetap ingin dipertahankan.
- Menyimpan sebagai insight — berguna jika outlier punya makna khusus, misalnya untuk mendeteksi fraud.

Karena jumlah outlier sedikit, saya memilih untuk menghapusnya agar analisis lebih akurat dan bersih.

### 1.  LOF

```python
import pandas as pd

# Baca dataset hasil deteksi LOF
df_lof = pd.read_csv("LOF.csv")

# Hapus outlier (Anomaly = 1)
df_lof_clean = df_lof[df_lof['Anomaly'] == 0]

print("Dataset LOF setelah outlier dihapus:")
print(df_lof_clean.to_string())
```
