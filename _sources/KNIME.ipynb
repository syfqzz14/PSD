# TOOLS KNIME Klasifikasi SMOTE

image.png

1. PostgreSQL Connector

      Node ini berfungsi untuk menghubungkan KNIME ke database PostgreSQL.
      Kamu memasukkan konfigurasi seperti:

      Host (localhost)

      Port (default 5432)

      Ecoli

      username (postgres)
      password (123)

2. DB Query Reader

    Setelah koneksi dibuat, node ini menjalankan query SQL untuk mengambil data dari database.
    Misalnya:

    SELECT * FROM Ecoli;


    Node ini menghasilkan output table berisi data mentah yang akan digunakan untuk analisis.

3. Column Filter

    Digunakan untuk memilih kolom yang relevan saja.
    Misalnya kamu hanya memilih fitur numerik dan kolom target seperti:

    mcg,gvh,lip,chg,aac,alm1,alm2


    Kolom yang tidak dibutuhkan (sequence_name) dibuang agar model lebih fokus.

4. Missing Value

    Node ini menangani nilai yang hilang (missing values).
    Kamu bisa mengatur bagaimana cara mengganti data kosong, misalnya:

    Mean/median untuk data numerik,

    Mode untuk data kategorik, atau

    Menghapus baris jika terlalu banyak nilai kosong.

5. Numeric Outliers

    Tahap ini mendeteksi nilai pencilan (outlier) dalam data numerik.
    Kamu bisa menandai atau menghapus data yang jauh dari distribusi normal
    Tujuannya agar model tidak bias oleh nilai ekstrem.

6. SMOTE

    Node ini menjalankan Synthetic Minority Over-sampling Technique (SMOTE) untuk menyeimbangkan dataset.
    Sehingga perlu diseimbangkan agar model tidak condong ke kelas mayoritas.

7. Table Partitioner

    Node ini membagi dataset menjadi dua bagian:

    Training set (misalnya 70%)

    Testing set (misalnya 30%)
    Digunakan agar model bisa belajar dan diuji pada data berbeda.

8. Naive Bayes Learner

    Node ini melatih model Naive Bayes menggunakan data training.
    Model akan mempelajari hubungan probabilistik antara fitur (mcg,gvh,lip,chg,aac,alm1,alm2) dan target (Outcome).

9. Naive Bayes Predictor

    Node ini menggunakan model yang telah dilatih untuk memprediksi kelas pada data testing.
    Output-nya berupa kolom baru (prediksi dan probabilitas tiap kelas).

10. Scorer

    Node ini menghitung akurasi dan metrik evaluasi lainnya seperti:

    Accuracy

    Precision

    Recall

    F1-score

    Confusion matrix

    Hasilnya menunjukkan seberapa baik model Naive Bayes dalam mengklasifikasikan data.

11. Bar Chart & Table View

    Node visualisasi ini digunakan untuk melihat hasil data dan performa model dalam bentuk grafik dan tabel:

    Sebelum SMOTE  untuk melihat distribusi kelas awal

    Sesudah SMOTE  untuk memastikan data sudah seimbang

    Setelah prediksi  untuk menampilkan hasil akhir

12. Numeric Outliers (Apply) + Python Script (Legacy)

    - Node ini merupakan percabangan tambahan untuk analisis lanjut:
      Numeric Outliers (Apply) digunakan untuk menerapkan penghapusan nilai outlier ke dataset baru.

    - Python Script (Legacy) bisa dipakai untuk menjalankan kode Python tambahan (misal analisis visual, heatmap korelasi, atau klasifikasi lain).
    Output-nya ditampilkan di Table View.
